```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_pickle('online-retail.pkl')
df
```

```python
df = df.drop(['InvoiceNo', 'StockCode', 'CustomerID'], axis=1)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df.duplicated().sum()
```

```python
df.info()
```

```python
df.InvoiceDate = pd.to_datetime(df.InvoiceDate, dayfirst=True)
df.info()
```

```python
df[['Description', 'Country']] = df[['Description', 'Country']].astype('category')
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
df.isnull().sum()
```

```python
df
```

```python
df.describe()
```

```python
df = df.query('Quantity >= 0 & UnitPrice >= 0').reset_index(drop=True)
```

```python
df
```

```python
df.describe()
```

```python
df_dates = df.groupby('InvoiceDate', as_index=False).sum(['Quantity', 'UnitPrice'])
df_dates
```

```python
df_dates.describe()
```

```python
px.line(df_dates, x='InvoiceDate', y='UnitPrice')
```
